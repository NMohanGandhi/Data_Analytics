```python
avg_max = new["Income"].max()//3
avg_max
```

```python
new["Income_Status"]= [i for i in range(1,len(new.index)+1)]
```

```python
new
```

```python
for i in range(len(new.index)):
    #print(new['Income'][i])
    if (new["Income"][i]>=1) & (new["Income"][i]<avg_max):
        #print(new["Income"][i],"is ok")
        new["Income_Status"][i]="Lower Class"
        
    elif (new["Income"][i]>=avg_max) & (new["Income"][i]<(avg_max + avg_max)):
        new["Income_Status"][i] = "Middle Class"
    else:
        new["Income_Status"][i] = "Upper Class"
        
    
    
```
