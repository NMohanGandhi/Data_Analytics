# A cosmetic company manufactures 4 different shades of lipstick. 
# The sale for 6 months is shown in the table. Represent it using bar charts.

```python
import matplotlib.pyplot as plt
```

```python
# Sales data
months = ['January', 'February', 'March', 'April', 'May', 'June']
shades = ['Shade 1', 'Shade 2', 'Shade 3', 'Shade 4']
sales_data = [
    [4500, 2870, 3985, 6855, 3200, 3456],
    [1600, 5645, 8900, 8976, 5678, 4555],
    [4400, 5675, 9768, 9008, 5643, 2233],
    [3245, 6754, 7786, 8965, 7865, 6547]
]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Lipstick Sales for Different Shades (6 months)')

for (i, j), ax in np.ndenumerate(axes):
    print(i,j)
    shade_index = 2 * i + j
    print(shade_index)
    ax.bar(months, sales_data[shade_index], color='skyblue')
    ax.set_title(f'Shade {shade_index + 1}')
    ax.set_xlabel('Months')
    ax.set_ylabel('Sales')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
