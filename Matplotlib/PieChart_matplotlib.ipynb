# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
cources = ["PythonFullStack","JavaFullStack","DataAnalytics"]
sizes =  [70,20,10]
```

```python
explode = (0,0,0)
```

```python
#labels = cources
```

```python
colors = ["b","g","r","c"]
```

```python
plt.pie(sizes,explode=explode,labels = cources,colors=colors,autopct = '%1.2f%%',shadow=True)
```

```python
explode = (0.1,0,0)
plt.pie(sizes,explode=explode,labels = cources,autopct = '%1.1f%%',shadow=True)
```

```python
explode = (0.4,0.3,0.2)
plt.pie(sizes,explode=explode,labels = cources,autopct = '%1.1f%%',shadow=True)
```

```python
explode = (0.1,0.1,0.1)
plt.pie(sizes,explode=explode,labels = cources,autopct = '%1.1f%%')
```
